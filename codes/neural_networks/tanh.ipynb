```python
# Following HIPS autograd https://github.com/HIPS/autograd/blob/master/examples/tanh.py
# Following https://github.com/QuantumBFS/SSSS/blob/master/1_deep_learning/tanh.py
import torch
%matplotlib inline
import matplotlib.pyplot as plt
x = torch.linspace(-7, 7, 100, requires_grad=True)
for i in range(7):
    if (i==0):
        y = torch.tanh(x/2)
    else:
        y, = torch.autograd.grad(y, x, grad_outputs=torch.ones(y.shape[0]), create_graph=True)
    plt.plot(x.detach().numpy(), y.detach().numpy(), '-', label='$%g$-th'%(i))
plt.legend(loc=2);plt.show()
```
